```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://ria.ru/'
response = requests.get(url).text # Загружаем страницу по ссылке
data = BeautifulSoup(str(response), 'html.parser') #Сохраняем html код страницы
print(data)
```

```python
len(data.find_all('a'))
```

```python
title0 = data.find('div', class_ ='cell-list__list').find('a', class_ ='cell-list__item-link color-font-hover-only')
link = title0.get('href')
article = title0.get('title')
print(article,":", link)
```

```python
#Собираем новости элемента div с классом "cell-list__item m-no-image" и ссылки на них

titles = data.find_all('div', class_ ='cell-list__item m-no-image')

for title1 in titles:
    link = title1.find('a', class_ ='cell-list__item-link color-font-hover-only').get('href')
    article = title1.find('a', class_ ='cell-list__item-link color-font-hover-only').get('title')
    print(article,":", link)
```
