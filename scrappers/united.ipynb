```python
import pandas as pd

df1 = pd.read_excel('parsed_news_kommersant_archive.xlsx')
df1 = df1.drop('Unnamed: 0', axis=1)

df2 = pd.read_excel('parsed_news_lenta_archive.xlsx')
df2 = df2.drop('Unnamed: 0', axis=1)

df3 = pd.read_excel('parsed_news_mkru_archive.xlsx')
df3 = df3.drop('Unnamed: 0', axis=1)

df4 = pd.read_csv('news_2610.csv', sep=';')
df4 = df4.drop('author', axis=1)
df5 = pd.read_csv('news_fontanka_1911.csv', sep=';')
```

```python
df4.rename(columns={'tags': 'tag'}, inplace=True)
df4.rename(columns={'date': 'date and time'}, inplace=True)
df4.rename(columns={'full_text': 'text'}, inplace=True)
```

```python
df5.rename(columns={'tags': 'tag'}, inplace=True)
df5.rename(columns={'date': 'date and time'}, inplace=True)
df5.rename(columns={'full_text': 'text'}, inplace=True)
```

```python
df_combined = pd.concat([df1, df2, df3, df4, df5], ignore_index=True)
df_combined
```

```python
df_combined.to_csv('parsed_news.csv', index=False, sep=';')
```
