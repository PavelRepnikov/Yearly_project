```python
import requests
from bs4 import BeautifulSoup
import re
from time import sleep
from tqdm import tqdm
import random
import pandas as pd
```

```python
url = 'https://lenta.ru/parts/news/1'
base_url = 'https://lenta.ru/parts/news/'
page_number = 1
urls = set()
```

```python
page = requests.get(url) 
soup = BeautifulSoup(page.text, 'html')
```

```python
len(soup.find_all('a'))
#page.status_code 
```

```python
needed_urls = set()
for link in soup.find_all('a'):
    if link.get('href') != None:
        if '/news/' in link.get('href'):
            needed_urls.add(link.get('href'))
            
```

```python
print(needed_urls)
len(needed_urls)
```

```python
while True:
    url = base_url + str(page_number)
    #print(url)

    #page = requests.get(url, timeout = 2.5) 
    page = requests.get(url) 
    
    if page.status_code != 200:
        break  
    #if page_number == 10000:
    #    break   
        
    soup = BeautifulSoup(page.text, 'html')
    
    prev_len = len(urls)
    
    for link in soup.find_all('a'):
        if link.get('href') != None:
            if not link.get('href').startswith('http'):
                if '/news/' in link.get('href'):
                    #needed_urls.add(link.get('href'))
                    full_url = 'https://lenta.ru' + link.get('href')
                    urls.add(full_url)
    page_number += 1   
    print(page_number)
    
    if prev_len == len(urls):
        break
    
    #sleep(random.random()) # задержка в 3 секунды
```

```python
# Используем регулярное выражение для фильтрации
filtered_urls = [url for url in urls if not re.match(r'https://lenta.ru/parts/news/\d+/', url)]

# Выводим отфильтрованный список
print(filtered_urls)
```

```python
def analyze_page(url0):
    page0 = requests.get(url0,)
    soup0 = BeautifulSoup(page0.text, 'html')
    
    date = str(soup0.find_all('a', {'class' : 'topic-header__item topic-header__time'}))
    soup1 = BeautifulSoup(date, "html.parser")
    time_tag = soup1.find("a", class_="topic-header__item topic-header__time")
    time_and_date = None
    if time_tag!=None:
        time_and_date = str(time_tag.text)

    tag = str(soup0.find_all('a', {'class' : 'topic-header__item topic-header__rubric'}))
    soup1 = BeautifulSoup(tag, "html.parser")
    rubric_tag = soup1.find("a", class_="topic-header__rubric")
    if rubric_tag:
        rubric_text = rubric_tag.text
    else: rubric_text = None
    
    text = soup0.find_all('p', {'class' : 'topic-body__content-text'})
    soup1 = BeautifulSoup(str(text), "html.parser")
    clean_text = soup1.get_text()
    
    return url0, time_and_date, rubric_text, clean_text
```

```python
news_lenta = [] # это будет список из кортежей, в которых будут храниться данные по каждой новости

for link in tqdm(filtered_urls):
    res = analyze_page(link)
    #print(res)
    news_lenta.append(res)
    #sleep(random.random()) # задержка в 3 секунды
```
